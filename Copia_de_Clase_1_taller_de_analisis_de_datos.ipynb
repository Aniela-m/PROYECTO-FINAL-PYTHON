#Flujo de control

```python
# IF, ELSE: Permite ejecutar codigo solo si se cumple una condicion
# Clasificar si una persona es menor o mayor edad
edad = 30
if edad >= 18:
  print("Es mayor de edad")
else:
    print("Es menor de edad")
```

```python
# While: ejecuta un bloque de codigo mientras se cumple una condicion.
contador = 0
print(f"contador inicial: {contador}")

while contador < 3:
  print(f"contador: {contador}")
  contador = contador + 1 # 0 --> 0 + 1 = 1 --> 1 + 1 = 2

print(f"contador final: {contador}")
```

```python
# Usa un for para sumar los numeros del 1 al 10
# 1 + 2 + 3 + 4 + 5 + ... + 10 = suma total
suma_total = 0

for i in range(1, 11):
  print(f"El numero en turno es: {i}")
  suma_total = suma_total + i

print(f"La suma total es: {suma_total}")

# suma_total = 0
# suma_total = 0 + 1 = 1
# suma_total = 1 + 2 = 3
# suma_ total = 3 + 4 = 7
```

```python
# La funcion input permite al usuario de introducir un valor de texto
palabra = input("Introduzca una palabra")
print(f"La palabra es: {palabra}")
```

```python
# Crea un while que cuente cuantas veces un usruario escribe algo antes de escribir la palabra "salir"
contador = 0
palabra = ""

while palabra != "salir":
  contador = contador + 1
  palabra = input("introduzca una palabra:")

print(f"numero de intentos: {contador}")
```

```python
# Crear una funcion que reciba un numero y muestre su tabla de multiplicador del 1 al 10
# 2 x 1 = 1
# 2 x 2 = 2
def tabla_multiplicar(numero):
  for i in range(1, 11):
    print(f"{numero} x {i} = {numero*i}")

tabla_multiplicar(12)
```

```python
# Pide tres calificaciones y muestra un promedio. Usa una funcion
def promedio(a, b, c):
  p = (a+b+c) / 3
  return p

c1 = float(input("Introduzca la calificacion 1:"))
c2 = float(input("Introduzca la calificacion 2:"))
c3 = float(input("Introduzca la calificacion 3:"))

try:
  print(f"El promedio final es de: {promedio(c1, c2, c3)}")
except:
  print("No se logro operar el promedio")
```
